```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/train.csv")
df["Target"] = (df["Heart Disease"] == "Presence") * 1
df = df.drop(["id", "Heart Disease"], axis=1)

print(df.info())
print(df.describe())
```

```python
CAT_COLS = ["Sex", "Chest pain type", "FBS over 120", "EKG results", "Exercise angina", "Slope of ST", "Number of vessels fluro", "Thallium"]
NUMERIC_COLS = ["Age", "ST depression", "BP", "Cholesterol", "Max HR"]
TARGET_COL = "Target"
for c in CAT_COLS:
    df[c] = df[c].astype("string")

for c in CAT_COLS:
    vc = df[c].value_counts().sort_index()
    plt.bar(vc.index, vc.values)
    plt.title(f"{c} - Value Counts")
    plt.xlabel(c)
    plt.ylabel("Count")
    plt.show()

for c in CAT_COLS:
    tr = df[[c, TARGET_COL]].groupby(c)[TARGET_COL].mean()
    plt.bar(tr.index, tr.values)
    plt.title(f"{c} - Target(=1) Rate")
    plt.xlabel(c)
    plt.ylabel("Target Rate ")
    plt.show()

for c in NUMERIC_COLS:
    plt.hist(df[c], bins=10)
    plt.title(f"{c} - Histogram")
    plt.xlabel(c)
    plt.ylabel("Frequency")
    plt.show()
    
c = TARGET_COL
vc = df[c].value_counts().sort_index()
plt.bar(vc.index, vc.values)
plt.title(f"{c} - Value Counts")
plt.xlabel(c)
plt.ylabel("Count")
plt.show()
```
